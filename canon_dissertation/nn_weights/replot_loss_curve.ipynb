```python
# Torus parameters
R = 5  # Major radius
r = 2  # Minor radius

# Create meshgrid for u and v
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, 2 * np.pi, 100)
U, V = np.meshgrid(u, v)

# Calculate x, y, z coordinates of the torus
X = (R + r * np.cos(V)) * np.cos(U)
Y = (R + r * np.cos(V)) * np.sin(U)
Z = r * np.sin(V)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface with transparency
ax.plot_surface(X, Y, Z, color='blue', alpha=0.4) # alpha between 0 (transparent) and 1 (opaque)


# Set labels and title (optional)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Transparent Toroid')
ax.set_aspect('equal', adjustable='box')
plt.tight_layout()

plt.show()
```
